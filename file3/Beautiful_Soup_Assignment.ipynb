Write a python program to display IMDB’s Top rated 100 Indian movies’ data
# **https://www.imdb.com/list/ls056092300/**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_imdb_list(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    movie_items = soup.find_all('div', class_='lister-item-content')

    names = []
    ratings = []
    years = []

    for movie in movie_items:
        name = movie.find('a').text.strip()
        names.append(name)

        rating = float(movie.find('span', class_='ipl-rating-star__rating').text.strip())
        ratings.append(rating)

        year = movie.find('span', class_='lister-item-year').text.strip('()')
        years.append(year)

    df = pd.DataFrame({'Name': names, 'Rating': ratings, 'Year': years})
    return df

url = "https://www.imdb.com/list/ls056092300/"
dataframe = scrape_imdb_list(url)
print(dataframe)
```


Write a python program to scrape first 10 product details which include product name , price , Image URL from

#https://www.bewakoof.com/bestseller?sort=popular

```python
import requests
from bs4 import BeautifulSoup

def scrape_bewakoof_bestsellers(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    product_cards = soup.find_all('div', class_='plp-product-card')
    products = []

    for card in product_cards[:10]:
        product_name = card.find(class_='h3-p-name').text.strip()
        price_actual = card.find(class_='actualPriceText').text.strip()
        image_url = card.find('img')['src']

        products.append({
            'Product Name': product_name,
            'Price': price_actual,
            'Image URL': image_url
        })

    return products

url = "https://www.bewakoof.com/bestseller?sort=popular%20."
bestsellers = scrape_bewakoof_bestsellers(url)
for product in bestsellers:
    print("Product Name:", product['Product Name'])
    print("Price:", product['Price'])
    print("Image URL:", product['Image URL'])
    print()
```

     
#  https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/

a) Paper title
b) date
c) Author

```python
import requests
from bs4 import BeautifulSoup

def scrape_most_downloaded_articles(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    article_listings = soup.find_all('div', class_='article-listing')
    articles_info = []

    for article in article_listings:
        title = article.find('h2', class_='article-title').text.strip()
        date = article.find('p', class_='article-date').text.strip()
        authors = article.find('p', class_='article-authors').text.strip()

        articles_info.append({
            'Title': title,
            'Date': date,
            'Authors': authors
        })

    return articles_info

url = "https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/"
articles = scrape_most_downloaded_articles(url)
for article in articles:
    print("Title:", article['Title'])
    print("Date:", article['Date'])
    print("Authors:", article['Authors'])
    print()
```
